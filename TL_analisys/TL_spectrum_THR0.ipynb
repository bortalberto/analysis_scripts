```python
h_sum=h_trans_hz_0.Clone()
```

```python
h_sum.Add(h_trans_hz_1)
# h_sum.Add(h_trans_hz_2)
h_sum.Add(h_trans_hz_3)
h_sum.Scale(0.333)
```

```python
h_sum.Draw("histo")
c.Draw()
```

```python
rate_real=h_sum.GetNbinsX()/3
N_bins=h_sum.GetNbinsX()
h_spec = R.TH1F( "h_spec", 'PSD vs Hz',int(N_bins/2) , 0, rate_real/2)
for n in range(0,int( N_bins/2)):
    h_spec.SetBinContent(n, (h_sum.GetBinContent(n) ** 2 / N_bins))
max=h_spec.GetMaximum()
for n in range(0, h_spec.GetNbinsX()):
    if (h_spec.GetBinContent(n) / max) !=0:
        h_spec.SetBinContent(n, 10*log10((h_spec.GetBinContent(n) / max)))
    else:
        h_spec.SetBinContent(n,-100)
```

```python
h_spec.Draw()
c.Draw()
```
