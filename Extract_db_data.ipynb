```python
 client = InfluxDBClient(host='localhost', port=8086)
```

```python
client.get_list_database()
```

```python
client.switch_database('GUFI_DB')
```

```python
result=client.query('SELECT mean("Humidity") FROM "Environment" WHERE time > now() - 12h  GROUP BY time(5s)',epoch='s')
```

```python
points=result.get_points()
```

```python
dict_4_pd={
    "time":[],
    "humidity":[]
}
for point in points:
    dict_4_pd["time"].append(point["time"]+28800)
    if point["mean"]==None:
            dict_4_pd["humidity"].append(0)
    else:
        dict_4_pd["humidity"].append(point["mean"])
```

```python
dict_4_pd
```

```python
humidity_pd=pd.DataFrame(dict_4_pd)
humidity_pd['time']=pd.to_datetime(humidity_pd['time'],unit='s')
```

```python
humidity_pd=humidity_pd.set_index("time")
```

```python
humidity_pd
```

```python
fig = go.Figure()
fig.add_trace( 
    go.Scatter(x=humidity_pd.index,y=humidity_pd.humidity,
              mode="lines")
)
fig.update_layout(title=dict(text='Signal hit/triggers during run 351', x=0.5),
                   yaxis_title='Signal hits/trigger',
                   xaxis_title='Time')
fig.show()
```

```python
client = InfluxDBClient(host='localhost', port=8086)
client.switch_database('GUFI_DB')
result=client.query('SELECT mean("Humidity") FROM "Environment" WHERE time > now() - 12h  GROUP BY time(5s)',epoch='s')
points=result.get_points()
dict_4_pd={
    "time":[],
    "humidity":[]
}
for point in points:
    dict_4_pd["time"].append(point["time"]+28800)
    if point["mean"]==None:
            dict_4_pd["humidity"].append(0)
    else:
        dict_4_pd["humidity"].append(point["mean"])
humidity_pd=pd.DataFrame(dict_4_pd)
humidity_pd['time']=pd.to_datetime(humidity_pd['time'],unit='s')
humidity_pd=humidity_pd.set_index("time")
```

```python
fig = go.Figure()
fig.add_trace( 
    go.Scatter(x=humidity_pd.index,y=humidity_pd.humidity,
              mode="lines")
)
fig.update_layout(title=dict(text='Signal hit/triggers during run 351', x=0.5),
                   yaxis_title='Signal hits/trigger',
                   xaxis_title='Time')
fig.show()
```
